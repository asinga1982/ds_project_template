```python
def analyze_client(clt):
    df1 = df[df.ClientCode == clt]
    df1_summ  = df1.groupby(["year","week_nbr"], as_index=False)["SampleCode"].count()
    pivoted = pd.pivot_table(df1_summ, values='SampleCode', columns='year', index='week_nbr')
    pivoted.plot(figsize=(20,6),kind="bar", );
    pivoted.reset_index(inplace=True)
    pivoted['Perc_change'] =  (pivoted[2018] - pivoted[2017])*100 /pivoted[2017] 
    return pivoted
```

```python
def weekly_avg(df1):
    df1 = df1[df1.ClientCode.isin(top_clients) & (df1.Arr_Date < '2019-04-01')]
    df1_summ  = df1.groupby(["year","week_nbr", "ClientCode"], as_index=False)["SampleCode"].count()
    return df1_summ.groupby("ClientCode", as_index=False)['SampleCode'].mean()
```

```python
#Weekly Avg for all date range
clt_weekly_avg = pd.DataFrame(weekly_avg(df1=df.copy()))
clt_weekly_avg.reset_index()
clt_weekly_avg.columns = ['ClientCode', 'Weekly_Avg']
clt_weekly_avg.head()
```

```python
#Generate combination of all clients and Weeks for Moving avg    
dfx = pd.DataFrame({'ClientCode':top_clients})
dfx['key'] = 1
dfy = pd.DataFrame({'week_nbr':[12,13,14,15,16,17,18]})
dfy['key'] = 1
df_all = pd.merge(dfx, dfy, on='key')[['ClientCode', 'week_nbr']]
df_all['SampleCount'] = 0
del dfx, dfy
df_all.head()
```

```python
#Calculate %Change from Weekly Avg
df2 = df[df.ClientCode.isin(top_clients) & (df.Arr_Date > '2019-03-31')].copy()
df2_summ  = df2.groupby(["week_nbr", "ClientCode"], as_index=False)["SampleCode"].count()
piv = pd.pivot_table(df2_summ, values="SampleCode",columns="week_nbr",index="ClientCode",fill_value=0)
piv.reset_index(inplace=True)

piv = pd.merge(piv, clt_weekly_avg,
                 on='ClientCode', 
                 how='left')

for i in range(14,19):
    piv[i] = (piv[i] - piv['Weekly_Avg'])*100/ piv['Weekly_Avg']

piv.columns = ['ClientCode', 'Week-4-WeeklyAvg', 'Week-3-WeeklyAvg', 'Week-2-WeeklyAvg', 'Week-1-WeeklyAvg', "Week-0-WeeklyAvg", 
               'Weekly_Avg']
print(piv.shape)
del df2_summ,df2
piv.head()
```

```python
#Calculate %Change from Moving Avg
df2 = df[df.ClientCode.isin(top_clients) & (df.Arr_Date < '2019-05-01') & (df.Arr_Date >= '2019-03-14')].copy()
df2_summ  = df2.groupby(["ClientCode", "year","week_nbr"], as_index=False)["SampleCode"].count()
df2_summ = pd.merge(df2_summ,df_all,how="right", on=["ClientCode","week_nbr"])
df2_summ.SampleCode.fillna(0, inplace=True)
df2_summ.drop('SampleCount', axis=1 ,inplace=True)
df2_summ.sort_values(by=["ClientCode", "week_nbr"], inplace=True)
df2_summ['new'] = df2_summ.groupby("ClientCode")["SampleCode"].rolling(window=4).mean().reset_index(0,drop=True)

df2_summ = df2_summ[df2_summ.week_nbr > 14]
pivi = pd.pivot_table(df2_summ, values='new', index='ClientCode', columns='week_nbr')
pivi.reset_index(inplace=True)

pivi.columns = ['ClientCode','Week3-MA4','Week2-MA4','Week1-MA4','Week0-MA4']
piv = pd.merge(piv, pivi, on="ClientCode", how="left")
del pivi, df2_summ

piv['Week3-MA4'] = (piv['Week3-MA4'] - piv['Weekly_Avg'])*100/ piv['Weekly_Avg']
piv['Week2-MA4'] = (piv['Week2-MA4'] - piv['Weekly_Avg'])*100/ piv['Weekly_Avg']
piv['Week1-MA4'] = (piv['Week1-MA4'] - piv['Weekly_Avg'])*100/ piv['Weekly_Avg']
piv['Week0-MA4'] = (piv['Week0-MA4'] - piv['Weekly_Avg'])*100/ piv['Weekly_Avg']

piv['Week2-MA4-change'] = piv['Week2-MA4'] - piv['Week3-MA4']
piv['Week1-MA4-change'] = piv['Week1-MA4'] - piv['Week2-MA4']
piv['Week0-MA4-change'] = piv['Week0-MA4'] - piv['Week1-MA4']

piv.head()
```

```python
#Change in YTD Volume from last year
df2 = df[df.ClientCode.isin(top_clients) & (df.Arr_Date < '2019-05-01') & (df.Arr_Date >= '2019-01-01')].copy()
df2_summ  = df2.groupby(["ClientCode"], as_index=False)["SampleCode"].count()

df3 = df[df.ClientCode.isin(top_clients) & (df.Arr_Date < '2018-05-01') & (df.Arr_Date >= '2018-01-01')].copy()
df3_summ  = df3.groupby(["ClientCode"], as_index=False)["SampleCode"].count()

df4 = df[df.ClientCode.isin(top_clients) & (df.Arr_Date < '2017-05-01') & (df.Arr_Date >= '2017-01-01')].copy()
df4_summ  = df4.groupby(["ClientCode"], as_index=False)["SampleCode"].count()

pivi = pd.merge(df2_summ, df3_summ, on='ClientCode', how="inner")
pivi = pd.merge(pivi, df4_summ, on='ClientCode', how="inner")

pivi.columns = ['ClientCode','SampleCount_2019','SampleCount_2018', "SampleCount_2017"]
pivi['YTD_Change_2018'] = (pivi.SampleCount_2019 - pivi.SampleCount_2018)*100/pivi.SampleCount_2018
pivi['YTD_Change_2017'] = (pivi.SampleCount_2019 - pivi.SampleCount_2017)*100/pivi.SampleCount_2017

pivi.YTD_Change_2018.fillna(0.00, inplace=True)
pivi.YTD_Change_2017.fillna(0.00, inplace=True)

pivi.drop(labels=["SampleCount_2019", "SampleCount_2018", "SampleCount_2017"], axis=1, inplace=True)

#pivi[pivi.ClientCode == "JK0000018"]

piv = pd.merge(piv, pivi,
                 on='ClientCode', 
                 how='left')
print(pivi.shape)

del pivi, df2_summ, df3_summ, df2, df3, df4, df4_summ
piv.head()
```

```python
# Change from same week last year
df1 = df[df.ClientCode.isin(top_clients) & (df.Arr_Date < '2019-05-01') & (df.Arr_Date >= "2018-01-01")]
max_week = df1.week_nbr[df1.Arr_Date > "2019-01-01"].max()
df1 = df1[df1.week_nbr.isin(range(max_week-4,max_week))]

df1_summ  = df1.groupby(["year","week_nbr", "ClientCode"], as_index=False)["SampleCode"].count()

pivi1 = pd.pivot_table(df1_summ[df1_summ.year==2019], index="ClientCode", columns="week_nbr", values="SampleCode", fill_value=0)
pivi2 = pd.pivot_table(df1_summ[df1_summ.year==2018], index="ClientCode", columns="week_nbr", values="SampleCode", fill_value=0)

pivi = (pivi1 - pivi2)*100/pivi2
pivi.fillna(0.00, inplace=True)
pivi.replace(np.inf, 100.00, inplace=True)
pivi.reset_index(inplace=True)
pivi.columns = ['ClientCode', 'Week4_YOY','Week3_YOY', 'Week2_YOY','Week1_YOY']
pivi['YoY_Total'] = pivi.Week1_YOY + pivi.Week2_YOY + pivi.Week3_YOY + pivi.Week4_YOY

piv = pd.merge(piv, pivi,
                 on='ClientCode', 
                 how='left')

del pivi, pivi1, pivi2,df1

piv.head()
#print(pivi1.shape)
#print(pivi2.shape)
```
